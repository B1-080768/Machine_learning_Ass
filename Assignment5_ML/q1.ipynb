# pca analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
df = pd.read_csv('wine_std.csv')
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
df.columns
```

```python
df
```

```python
wine_pca = PCA(n_components = 2)
wine_pca.fit(df)
w = wine_pca.transform(df)
w1 = pd.DataFrame(w)
w1.columns = ['pc1','pc2']
w1
```

```python
# wine_pca.n_components_
df = pd.concat([df,w1],axis = 1)
df
```

```python
from sklearn.cluster import AgglomerativeClustering
ac2 = AgglomerativeClustering(n_clusters = 3)
 
# Visualizing the clustering
plt.figure(figsize =(6, 6))
plt.scatter(df['pc1'],df['pc2'], 
           c = ac2.fit_predict(df), cmap ='rainbow')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.title('visualisation of wine data using diamentoinality reduction')
plt.show()
```
