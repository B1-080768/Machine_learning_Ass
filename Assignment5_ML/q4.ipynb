```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('crime_std.csv')
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
df1  = pd.DataFrame(df)
```

```python
df2 = df[['Murder','Assault','Rape']]
```

```python
pca = PCA(n_components = 1)
pca.fit(df2)
p1 = pca.transform(df2)
p2 = pd.DataFrame(p1)
p2.columns = (['pc1'])
p2
df1  = pd.concat([df1,p2],axis = 1)
df1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac  = AgglomerativeClustering(n_clusters =3)
plt.figure(figsize=(6,6))
plt.scatter(df1['UrbanPop'],df1['pc1'],c = ac.fit_predict(df1),cmap = 'rainbow')
plt.xlabel('UrbanPop')
plt.ylabel('pc1')
plt.title('urban population vrs crime data')
plt.show()
```
