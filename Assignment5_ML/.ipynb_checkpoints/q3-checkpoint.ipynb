### pca of universities data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('universities_std.csv')
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
```

```python
pca = PCA(n_components = 1)
pca.fit(df)
p1 = pca.transform(df)
p2 = pd.DataFrame(p1)
p2.columns = ['pc1']
p2
df = pd.concat([df,p2],axis = 1)
df
```

### representation of data

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters = 3)
plt.figure(figsize = (6,6))
plt.scatter(df['SAT'],df['pc1'],c=ac.fit_predict(df),cmap = 'rainbow')
plt.xlabel('SAT')
plt.ylabel('pc1')
plt.title("visualization of universities data")
plt.show()
```
