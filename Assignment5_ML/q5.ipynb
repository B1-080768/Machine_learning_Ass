```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
df = pd.read_csv("Grocery_std.csv")
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
df.columns
```

```python
df1 = df[['Fresh', 'Milk', 'Frozen', 'Detergents_Paper', 'Delicassen']]
```

```python
pca = PCA(n_components = 1)
pca.fit(df1)
p1 = pca.transform(df1)
p2 = pd.DataFrame(p1)
p2.columns= ['pc1']
p2
df = pd.concat([df,p2],axis = 1)
df
```

```python
from sklearn.cluster import AgglomerativeClustering
ac  = AgglomerativeClustering(n_clusters = 2)
plt.figure(figsize = (6,6))
plt.scatter(df['Grocery'],df['pc1'],c = ac.fit_predict(df),cmap = 'rainbow')
plt.xlabel('Grocery')
plt.ylabel('pc1')
plt.title('visualization of grocery data ')
plt.show()
```
