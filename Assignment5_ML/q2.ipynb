```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Framingham_std.csv')
df.drop('Unnamed: 0',axis = 1 ,inplace = True)
```

```python
### pca of data
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(df)
p = pca.transform(df)
p1 = pd.DataFrame(p) 
p1.columns = ['pc1','pc2']
df = pd.concat([df,p1],axis  =1 )
```

```python
### representation of data 
```

```python
from sklearn.cluster import AgglomerativeClustering
ac  = AgglomerativeClustering(n_clusters = 3)
plt.figure(figsize = (6,6))
plt.scatter(df['pc1'],df['pc2'],c = ac.fit_predict(df),cmap = 'rainbow')
plt.xlabel('pc1')
plt.ylabel('pc2')
plt.title("visualization of framingham data ")
plt.show()
```
