```python
#  problem defination Que2.-  
# In this dataset you have to predict that whether we have to give ATTORNEY to the clients or not (0=Yes, 1=No) based on below factors.
# CLMAGE - Age of Client
# CLMSEX - Sex of client (0=Male, 1= Female)
# SEATBELT - Whether client has Seatbelt or not (0=Yes, 1=No)
# CLMINSUR - Whether Client give Insurance or not (0=Yes, 1=No)
# LOSS - Loss that client faced
# Prepare model by using Decision Tree, Random Forest, CatBoost, XGBoost & 
# Also calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score,precision, recall.
# Dataset : Claimants.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('claimants.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
```

```python
df1.fillna(0)
```

```python
df.corr()
```

```python
X = df1.drop(['ATTORNEY','CLMSEX','CLMINSUR','CLMAGE'],axis = 1)
Y = df1[['ATTORNEY']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,train_size = 0.7,random_state = 12345)
```

```python
def build_dt_model():
       from sklearn.tree import DecisionTreeClassifier
       model = DecisionTreeClassifier()
       model.fit(x_train,y_train)
       return model
     
```

```python
def build_rf_model():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def build_cb_model():
    from catboost import CatBoostClassifier
    model = CatBoostClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def build_xgb_model():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
# evaluate module
def evaluate_model(model):
    y_true = y_test
    y_pred = model.predict(x_test)
    from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
    accuracy = f"{accuracy_score(y_true,y_pred):.2f}"
    precision = f"{precision_score(y_true,y_pred):.2f}"
    recall = f"{recall_score(y_true,y_pred):.2f}"
    f1_score = f"{f1_score(y_true,y_pred):.2f}"
    return accuracy,precision,recall,f1_score
```

```python
# create module and evaluate them
evaluation_report = []
model_dt = build_dt_model()
evaluation_report.append({"model":"decision tree ","metrics":{evaluate_model(model_dt)}})
model_rf = build_rf_model()
evaluation_report.append({"model":"random forest ","metrics":{evaluate_model(model_rf)}})
model_cb = build_cb_model()
evaluation_report.append({"model":"catboost","metrics":{evaluate_model(model_cb)}})
model_xgb = build_xgb_model()
evaluation_report.append({"model":"xg boost ","metrics":{evaluate_model(model_xgb)}})
```

```python
print(f"{'Algorithm':<20} => 'Accuracy', 'precision', 'recall', 'f1'")
print()
for info in evaluation_report:
    print(f"{info['model']:<20} => {info['metrics']}")
```
