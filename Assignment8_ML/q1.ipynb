```python
# Que.1 -
# In this dataset you have to predict that customer is excited or not based on various factors given in dataset.
# Dataset : Churn.csv
# Steps to follow:
# 1. Do Proper EDA.
# 2. Drop Unnecessary Columns.
# 3. convert Categorical columns into numerical columns.Prepare model by using Decision Tree, Random Forest, CatBoost, 
# XGBoost & Also calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score,precision, 
# recall
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
df = pd.read_csv('Churn.csv')
df.head(3)
```

```python
df.drop(['RowNumber','CustomerId','Surname','Geography'],axis = 1,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

# create an encoder object
encoder = LabelEncoder()

# fit the data
encoder.fit(df['Gender'])

# transform the Gender column
df['Gender'] = encoder.transform(df['Gender'])
df
```

```python
X = df.drop(['Exited','HasCrCard','CreditScore','Tenure','NumOfProducts','EstimatedSalary'],axis = 1)
Y = df[['Exited']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, random_state=12345)
```

```python
def build_dt_model():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x_train, y_train)
    return model
    
```

```python
def build_rf_model():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=100)
    model.fit(x_train, y_train) 
    return model
```

```python
def build_cb_model():
    from catboost import CatBoostClassifier
    model = CatBoostClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def build_xgb_model():
    from xgboost import XGBClassifier
    
    # create a model
    model = XGBClassifier()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def evaluate_model(model):
    y_true = y_test
    y_pred = model.predict(x_test)
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    accuracy = f"{accuracy_score(y_true, y_pred):.2f}"
    precision = f"{precision_score(y_true, y_pred):.2f}"
    recall = f"{recall_score(y_true, y_pred):.2f}"
    f1 = f"{f1_score(y_true, y_pred):.2f}"

    return accuracy, precision, recall, f1
```

```python
evaluation_report = []

model_dt = build_dt_model()
evaluation_report.append({"model": "Decision Tree", "metrics": evaluate_model(model_dt)})

model_rf = build_rf_model()
evaluation_report.append({"model": "Random Forest", "metrics": evaluate_model(model_rf)})

model_cb = build_cb_model()
evaluation_report.append({"model": "CatBoost", "metrics": evaluate_model(model_cb)})

model_xgb = build_xgb_model()
evaluation_report.append({"model": "XGB", "metrics": evaluate_model(model_xgb)})
```

```python
print(f"{'Algorithm':<20} => 'Accuracy', 'precision', 'recall', 'f1'")
print()
for info in evaluation_report:
    print(f"{info['model']:<20} => {info['metrics']}")
```
