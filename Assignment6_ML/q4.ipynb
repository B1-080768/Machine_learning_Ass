```python
# Que. 4
# In the following dataset you have given SAT score and GPA of students. You have to predict GPA of students by using SAT score.
# Build a Linear Regression Model and Calculate different errors to evaluate the model.
# Dataset : SAT-GPA.csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('SAT-GPA.csv')
df.head(3)
```

```python
X = df.drop('GPA',axis = 1)
Y = df[['GPA']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
plt.scatter(df['SAT'],df['GPA'])
y_predict = model.predict(X)
plt.plot(df['SAT'],y_predict,c = "red",label = "regression line")
plt.scatter(df['SAT'],y_predict,c = "red")
plt.title("sat vrs gpa")
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

```python
### split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.7,random_state =12345)
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_true = y_test
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_true,y_pred)
mse = mean_squared_error(y_true,y_pred)
rsme = np.sqrt(mse)
r2_score = r2_score(y_true,y_pred)
r2_score
print(f"mae = {mae},mse = {mse},rsme = {rsme},r2 = {r2_score}")
```
