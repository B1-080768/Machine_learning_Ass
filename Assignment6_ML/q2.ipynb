```python
# Que. 2 In the given dataset TVMarketing.csv you have given  sales based on 'TV' marketing budget.
# Build a linear regression model to predict 'Sales' using 'TV' as the predictor variable.
# Also Evaluate the model by using different methods.
# Dataset : TVMarketing.csv
### import all the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv('TvMarketing.csv')
df.head(3)
```

```python
X = df.drop('Sales',axis = 1) ### data preprocessing
Y = df[['Sales']]
```

```python
model = LinearRegression()  ### create a model
model.fit(X,Y)
plt.scatter(df['TV'],df['Sales'])
y_predict = model.predict(X)
plt.plot(df['TV'],y_predict,c = "red",label = "regression line")
plt.scatter(df['TV'],y_predict,c = "red")
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('tv vrs sales')
plt.show()
```

### evaluate the model

```python
from sklearn.model_selection import train_test_split ### split the data 
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.7,random_state = 12345)
model = LinearRegression() ## model creation
model.fit(x_train,y_train)
y_true = y_test
y_pred = model.predict(x_test)
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_true,y_pred)
mae
mse = mean_squared_error(y_true,y_pred)
mse
rmse = np.sqrt(mse)
rmse
r2_score = r2_score(y_true,y_pred)
r2_score
print(f"mae ={mae:.2f},mse = {mse:.2f},rmse = {rmse:.2f},r2_score = {r2_score:.2f}")
y_pred
```
